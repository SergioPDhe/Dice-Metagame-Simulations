```python
import os
import sys
import random
import math
import pprint
import timeit

faceset = list()

ranks = {}
movement = {}

K = 10
```

```python
def Probability(rating1, rating2): 
    return 1.0 * 1.0 / (1 + 1.0 * math.pow(10, 1.0 * (rating1 - rating2) / 400))
```

```python
#def generate():
#    for a in range(0,22):
#        for b in range(0,22):
#            for c in range(0,22):
#                 for d in range(0,22):
#                     for e in range(0,22):
#                         for f in range(0,22):
#                             if a+b+c+d+e+f==21:
#                                 facelist = sorted(list([a,b,c,d,e,f]))
#                                 if facelist not in faceset:
#                                     faceset.append(facelist)
                                    
def generate():# every [possible 6-sided die with 21 total]
    for a in range(4,22):# largest number in the dice can only be between 4 and 21 ([4,4,4,3,3,3] to [21,0,0,0,0,0])
        for b in range(11):# second largest number can only be as high as 10 [11,10,0,0,0,0]
            if a<b:
                break # if the 2nd number is greater than the 1st, quit out of loop
            for c in range(8):# 3rd number goes up to 7 [7,7,7,0,0,0]
                if b<c:
                    break
                for d in range(6):# 4th number goes to 5 [6,5,5,5,0,0]
                    if c<d:
                        break
                    for e in range(5):# and so on [5,4,4,4,4,0]
                        if d<e:
                            break
                        for f in range(4):# [4,4,4,3,3,3]
                            if e<f:
                                break
                            if (a+b+c+d+e+f)==21:# only add to list if numbers add to 21
                                faceset.append(list([a,b,c,d,e,f]))
```

```python
def setranks():
    for a in faceset:
        ranks[str(a)] = 1500
        movement[str(a)] = 0
```

```python
def runround(n):
    for i in range(n):
        for a in faceset:
            for b in faceset:
                for c in a:
                    for d in b:
                        Pb = Probability(ranks[str(a)], ranks[str(b)])
                        Pa = Probability(ranks[str(b)], ranks[str(a)]) 
                        if c>d:
                            movement[str(a)] = movement[str(a)] + K * (1 - Pa) / 36
                            movement[str(b)] = movement[str(b)] + K * (0 - Pb) / 36
                        elif c<d:
                            movement[str(a)] = movement[str(a)] + K * (0 - Pa) / 36
                            movement[str(b)] = movement[str(b)] + K * (1 - Pb) / 36
                        elif c==d:
                            movement[str(a)] = movement[str(a)] + K * (0.5 - Pa) / 36
                            movement[str(b)] = movement[str(b)] + K * (0.5 - Pb) / 36
        for e in faceset:
            ranks[str(e)] = ranks[str(e)] + movement[str(e)]
            movement[str(e)] = 0
```

```python
generate()

setranks()
```

```python
runround(1)

pprint.pprint(sorted(ranks.items(), key=lambda x: x[1]))
```
